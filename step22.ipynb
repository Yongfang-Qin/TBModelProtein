```python
import numpy as np
import math
```

```python
def pointDistance(pointFir, pointSec):
    pointFir = np.array(pointFir)
    pointSec = np.array(pointSec)
    diff = abs(pointFir - pointSec)
    dist = math.sqrt(sum(diff * diff.T))
    return dist
```
